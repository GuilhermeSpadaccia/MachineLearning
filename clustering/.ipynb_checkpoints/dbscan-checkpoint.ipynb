```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
eps = 3
minpoints = 4
```

```python
df = pd.read_csv("dataset.txt",  delimiter="\t")
```

```python
plt.scatter(df.x.values, df.y.values, c=df['oriclass'])
plt.show()
```

```python
df['setofpoints'] = 'unclassified'
df['neighbors'] = 0
df['neiborhood'] = None
df['dataclass'] = 0
```

```python
df
```

```python
def manhattan_dist(M, P):
    return np.abs(M[0]-P[0]) + np.abs(M[1]-P[1])
    
```

```python
def get_neighbors(i, data_class, reachable=False):
    print("Porcessando ponto: ",i)
    
    neiborhood = []
    neighbors = 0
    update_class = False
    
    for x in range(len(df)):
        if i != x:
            # get M point
            M = [df.iloc[i]['x'], df.iloc[i]['y']]
            # get P point
            P = [df.iloc[x]['x'], df.iloc[x]['y']]
            # calculate the Manhattan distance from point M to P
            mdist = manhattan_dist(M,P)
            # Count the neiborhood
            if mdist <= eps:
                neighbors += 1
                neiborhood.append(x)
    
    df['neighbors'][i] = neighbors
    df['neiborhood'][i] = neiborhood
    
    if neighbors == 0:
        df['setofpoints'][i] = 'noise'
    if 0 < neighbors < minpoints:
        df['setofpoints'][i] = 'border'
        if reachable:
            df['dataclass'][i] = data_class
    if neighbors >= minpoints:
        df['setofpoints'][i] = 'core'
        df['dataclass'][i] = data_class
        update_class = True
        
        #print("neiborhood ->", neiborhood)
        for k in neiborhood:
            #print("-- RECURSAO - VALOR DE DATACLASS DE K ---> ",df['dataclass'][k])
            if df['dataclass'][k] == 0:
                #print("Ponto atual: {} - Chamando recurso para o ponto: {}".format(i,k))
                update_class = get_neighbors(k, data_class, True)
        
    print("Point {} -> {} have {} neighbors: {}".format(i, M, neighbors, neiborhood))
    
    return update_class
```

```python
# select a data point M to be analysed
data_class = 1
for i in range(len(df)):
    # Compare M to all other data points
    if df['setofpoints'][i] == 'unclassified':
        update_class = get_neighbors(i, data_class)
    
    if update_class:
        print("MUDOU VALOR PARA i = ",i)
        data_class += 1
        update_class = False
```
